## Zadanie 3

Stworz funkcję która usunie "wyplaszczy" zagnieżdzone listy

l = [4, 1, [ 2, [7] ], 3]

wyplaszcz(l)

l

[4, 1, 2, 7, 3]

````python
def wyplaszcz(lista):
    ret = []
    for x in lista:
        if type(x)==type([]):
            for y in wyplaszcz(x):
                ret.append(y)
        else:
            ret.append(x)
    return ret
````

````python
l = [4, 1, [ 2, [7] ], 3]
l = wyplaszcz(l)
l
````

````python
def wyplaszcz(lista):
    kopia = list(lista)
    for i in range(len(lista)-1,-1,-1):
        del lista[i]
    for x in kopia:
        if type(x)==type([]):
            for y in wyplaszcz(x):
                lista.append(y)
        else:
            lista.append(x)
    return lista
````

````python
l = [4, 1, [ 2, [7] ], 3]
wyplaszcz(l)
l
````

````python
def color_negative_red(val):
    color = 'red' if val < 0 else 'white'
    return 'color: %s' % color
````

````python
color_negative_red(1)
````

````python
color_negative_red(-1)
````

````python
df.style.applymap(color_negative_red)
````

````python
df.index = pd.to_datetime(df.index, format = '%Y-%m-%d').strftime('%Y-%m-%d')   # formatowanie indeksu
df.style.applymap(color_negative_red)
````

````python
def highlight_max(s):
    is_max = s == s.max()
    return ['background-color: blue' if v else '' for v in is_max]
````

````python
df.style.apply(highlight_max)   # w każdej kolumnie pokazuje max (domyślnie jest tutaj ukryte axis 0 - to kolumny)
````

````python
df.style.apply(highlight_max, axis=0)   # to samo co wyżej widać
````

````python
df.style.apply(highlight_max, axis=1)   # oś 1 to wiersze
````

Łączymy te 2 rzeczy:

````python
df.style.apply(highlight_max, axis=0).applymap(color_negative_red)
````

Wolny sposób poniżej! Iteracja przez obiekty pandasa jest generalnie wolna

````python
# chcemy dodać kolumnę ratio

data = []

for row in df.iterrows():
    x = row[1][0]
    y = row[1][1]
    data.append([x, y, x/y])
    
df2 = pd.DataFrame(data, columns = ['x', 'y', 'ratio'])
df2
````

Lepiej - <b>list comprehension:

````python
df3 = pd.DataFrame(
    ((row[1][0], row[1][1], row[1][0]/row[1][1]) for row in df.iterrows()), columns = ["x","y","ratio"]
)
df3
````

Jeszcze lepiej - <b> apply:

````python
# tworzymy kopię df
df4 = pd.DataFrame.copy(df)
df4
````

````python
def ratio(row):
    return row.x/row.y

df4['ratio'] = df4.apply(ratio, axis = 1)
````

````python
df4
````

Najlepiej - <b> vectorization </b> (traktujemy kolumny jako wektory i robimy na nich operacje):

````python
df5 = pd.DataFrame.copy(df)
df5
````

````python
df5['ratio'] = df5.x/df5.y
df5
````
